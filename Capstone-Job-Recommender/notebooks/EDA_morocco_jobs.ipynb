```python
import pandas as pd

csv_path = datasets_folder_path/"raw/morocco_jobs_sampled (5).csv"
df_m = pd.read_csv(csv_path, low_memory=False)
df_m["location"] = df_m["location"] + ", " + df_m["Country"]
df_m.drop(columns=["Country"], inplace=True)
print(df_m.head())

# display only column names
print("\nColumn list:")
print(list(df_m.columns), "\n")
```

Basic Inspection for Dataset

```python
# Shape of the DataFrame
print("Shape (rows, cols):", df_m.shape, "\n")

# Data types of each column
print("Data types:")
print(df_m.dtypes, "\n")

# Missingness summary (count + percent)
missing_summary = (
    df_m.isna()
      .sum()
      .to_frame(name="num_missing")
      .assign(pct_missing=lambda x: (x["num_missing"] / len(df_m)).round(4))
      .sort_values("pct_missing", ascending=False)
)
print("Missingness summary (top 12):")
print(missing_summary.head(12), "\n")

# Fully duplicate rows
print("Fully duplicated rows:", df_m.duplicated().sum())

# Quick preview
display(df_m.head(20))
```

Step 2.1 — Unique values distribution (Job Title, Role, Work Type)

```python
# 2.1 — Unique values distribution
target_cols = ["Job Title", "Role", "Work Type"]

for col in target_cols:
    if col not in df_m.columns:
        print(f"[WARN] Column missing: {col}")
        continue
    
    s = df_m[col].astype(str).str.strip()
    n_unique = s.nunique(dropna=True)
    print(f"\n=== {col} ===")
    print(f"Unique categories: {n_unique}")
    
    # Frequency table (top 20)
    vc = s.value_counts(dropna=False)
    print("\nTop 20 frequencies:")
    print(vc.head(20).to_string())
    
    # (Optional) show number of singletons (appearing once)
    singletons = (vc == 1).sum()
    print(f"\nSingleton categories (appear exactly once): {singletons}")
```

Step 2.2 — Duplicates in “signal columns” (Role, Job Title, skills)

```python
# 2.2 — Duplicates in signal columns
signal_cols = ["Role", "Job Title", "skills"]

# Check availability
missing = [c for c in signal_cols if c not in df_m.columns]
if missing:
    print(f"[WARN] Missing signal columns: {missing}")
else:
    # Normalize for fair comparison (trim only; keep case as-is to avoid over-merging)
    key_df_m = df_m[signal_cols].astype(str).apply(lambda s: s.str.strip())
    
    dup_mask = key_df_m.duplicated(keep="first")
    num_dup_rows = int(dup_mask.sum())
    total_rows = len(df_m)
    
    collapsed = df_m.loc[~dup_mask].copy()
    collapsed_rows = len(collapsed)
    collapsed_pct = round(100 * (total_rows - collapsed_rows) / total_rows, 2)
    
    print("\n=== Duplicate Analysis (Role, Job Title, skills) ===")
    print(f"Total rows: {total_rows}")
    print(f"Duplicate rows (same triplet): {num_dup_rows}")
    print(f"Rows after dropping duplicates: {collapsed_rows}")
    print(f"Percent rows removed by collapsing: {collapsed_pct}%")
    
    # (Optional) inspect a few duplicate groups
    print("\nSample duplicate groups (up to 5 groups, 3 rows each):")
    dup_groups = (
        df_m[dup_mask | key_df_m.duplicated(keep=False)]
        .sort_values(signal_cols)
        .groupby(signal_cols, as_index=False, group_keys=True)
        .head(2)
    )
    # Limit to ~5 groups * 3 rows = 15 rows for display
    try:
        from itertools import islice
        # Get first 5 groups
        first_groups_keys = list(islice(
            (g for g, _ in df_m.groupby(signal_cols) if len(df_m.loc[(key_df_m==list(g)).all(axis=1)])>1),
            10
        ))
    except Exception:
        first_groups_keys = []
    
    # Simpler cap:
    print(
        dup_groups[signal_cols + [c for c in ["Company", "location", "Work Type"] if c in df_m.columns]]
        .head(25)
        .to_string(index=False)
    )
    
    # collapsed is your deduplicated dataframe if you need it later:
    collapsed.to_csv("morocco_dedup.csv", index=False)
```

Step 2.3 - Text sanity checks (length stats for skills / description / responsibilities)

```python
# STEP 2.3 — text length stats
text_cols = [c for c in ['skills', 'Job Description', 'Responsibilities'] if c in df_m.columns]

def text_len_report(frame, cols):
    out = {}
    for col in cols:
        s = frame[col].astype(str).fillna("")
        lens = s.str.len()
        out[col] = {
            "num_rows": len(s),
            "num_missing": frame[col].isna().sum(),
            "num_empty": int((s.eq("") | s.str.strip().eq("")).sum()),
            "min_len": int(lens.min()),
            "max_len": int(lens.max()),
            "mean_len": float(lens.mean()),
            "median_len": float(lens.median()),
            "pct_empty": round(100 * (s.eq("") | s.str.strip().eq("")).mean(), 2),
        }
    return pd.DataFrame(out).T.sort_values("pct_empty", ascending=False)

text_stats = text_len_report(df_m, text_cols)
print("=== Text Column Length Stats ===")
print(text_stats.to_string())
```

Step 2.4 — Numeric salary checks (+ quick consistency peek vs range)

```python
# STEP 2.4 — salary numeric checks
import numpy as np

if 'Salary_Numeric' in df_m.columns:
    s = pd.to_numeric(df_m['Salary_Numeric'], errors='coerce')
    salary_stats = {
        "num_missing": int(s.isna().sum()),
        "num_zero": int((s.fillna(0) == 0).sum()),
        "min": float(np.nanmin(s)),
        "max": float(np.nanmax(s)),
        "mean": float(np.nanmean(s)),
        "median": float(np.nanmedian(s)),
        "num_unique": int(s.nunique(dropna=True)),
    }
    print("=== Salary_Numeric Stats ===")
    for k, v in salary_stats.items():
        print(f"{k}: {v}")

    # quick consistency peek
    if 'Salary Range' in df_m.columns:
        print("\n=== Salary Range vs Salary_Numeric (sample 20) ===")
        print(df_m[['Salary Range','Salary_Numeric']].head(20).to_string(index=False))
else:
    print("Column 'Salary_Numeric' not found.")
```
