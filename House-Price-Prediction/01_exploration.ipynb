#### Checking normality & skewness of numerical columns
 
- We handle outliers of numerical columns. 
- The concept of an outlier for numerical data is based on being an extreme value far from the central tendency (mean/median). For categorical data, an outlier is an observation that is unusually infrequent or inconsistent with the other categories.

✅ 1. Visual Methods (Quick Insight)

```python
import matplotlib.pyplot as plt
import seaborn as sns

numeric_cols = ['price', 'area', 'bedrooms', 'bathrooms', 'stories', 'parking']

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

✅ 2. QQ Plot (Quantile-Quantile Plot)

```python
import scipy.stats as stats
import matplotlib.pyplot as plt

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    stats.probplot(df[col], dist="norm", plot=plt)
    plt.title(f'Q-Q plot of {col}')
    plt.show()
```

✅ 3. Shapiro-Wilk Test (Statistical Test for Normality)

```python
from scipy.stats import shapiro

for col in numeric_cols:
    stat, p = shapiro(df[col])
    print(f'{col}: p-value = {p:.4f}')
```

## Analysis of Categorical Columns

### ✅ 1. Countplots for All Categorical Columns
Shows class imbalance in categorical features (e.g., more houses near mainroad?)
How it helps:
- You’ll identify skewed features that may need balancing or re-coding.

```python
import seaborn as sns
import matplotlib.pyplot as plt

categorical_cols = df.select_dtypes(include='category').columns

for col in categorical_cols:
    plt.figure(figsize=(6, 3))
    sns.countplot(data=df, x=col, palette="Set2")
    plt.title(f"Countplot of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```
